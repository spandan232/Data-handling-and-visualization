```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
```

```python
text_data= "Python is a popular programming language used for data and is analysis and visualization technique I love Python"
wordcloud = WordCloud(width=800,height=400, background_color='black').generate(text_data)
```

```python
plt.figure(figsize=(10,9))
plt.imshow(wordcloud)
plt.title('Word Cloud')
plt.show()
```

```python
words=text_data.split()
word_counts=Counter(words)
plt.figure(figsize=(9,5))
plt.bar(word_counts.keys(),word_counts.values())
plt.title('Histogram of Word Frequecies')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```
